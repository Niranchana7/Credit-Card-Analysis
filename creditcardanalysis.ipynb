```python
features = ["Unnamed: 0","X1", "X23"]   # target features
x=data[features] 
print(x.head())
```

```python
y=data.Y
print(y.head())
```

```python
from sklearn.tree import DecisionTreeRegressor     #pediction of target variables
model=DecisionTreeRegressor(random_state=1)
model.fit(x,y)

predict=model.predict(x)

print(predict)
```

```python
from sklearn.metrics import mean_absolute_error

mae=mean_absolute_error(y,predict)
print(mae)
```

```python
from sklearn.model_selection import train_test_split
    
train_x,val_x,train_y,val_y=train_test_split(x,y,random_state=0,test_size=0.4)

model=DecisionTreeRegressor()

model.fit(train_x,train_y)

predict=model.predict(val_x)

print(mean_absolute_error(val_y,predict))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(random_state=1)

model.fit(train_x,train_y)

predict=model.predict(val_x)
    
x=mean_absolute_error(val_y,predict)

print(x)
```
